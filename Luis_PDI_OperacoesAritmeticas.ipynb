```python
img2 = cv2.imread('vegeta.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2.shape
```

```python
plt.imshow(img2)
```

```python
img_red = cv2.resize(img2, (750,1000), interpolation = cv2.INTER_AREA)
```

```python
img_red.shape
```

```python
fig = plt.figure(figsize=(10, 10))

plt.subplots_adjust(left=0.125, bottom=0.1, right=1.5, top=1,
                        wspace=0.12, hspace=0.1)

x1_plt = fig.add_subplot(2, 2, 1)
x2_plt = fig.add_subplot(2, 2, 2)

x1_plt.axis("off")
x1_plt.set_title("Imagem 1")
x1_plt.imshow(img1)
x2_plt.axis("off")
x2_plt.set_title("Imagem 2")
x2_plt.imshow(img_red)
plt.show()
```

**SOMA**

```python
imgsomada = cv2.add(img_red,img1)
```

```python
fig = plt.figure(figsize=(20,5))
plt.imshow(imgsomada)
plt.axis('off')
plt.title("Resultado da Adio")
```

**SUBTRACO**

```python
imgsubtraida = cv2.subtract(img_red,img1)
```

```python
fig = plt.figure(figsize=(20,5))
plt.imshow(imgsubtraida)
plt.axis('off')
plt.title("Resultado da Subtrao")
```

**MISTURA**

```python
imgTotal = cv2.addWeighted(img1,0.8, img_red, 0.8,0)
```

```python
fig = plt.figure(figsize=(20,5))
plt.imshow(imgTotal)
plt.axis('off')
plt.title("Resultado da mesclagem")
```
